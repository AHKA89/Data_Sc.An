### SIMPLE LINEAR REGRESSION

#### 1. Global Sales vs Sales in North America

```python
 # Extract features
x=video_df_new["NA_Sales"]
y=video_df_new["Global_Sales"]

 # Reshape x to have 2 dimensions
x=x.to_numpy().reshape(-1,1)

 # Split into training and testing datasets
from sklearn.model_selection import train_test_split

 #Use 80% for training 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) 
```

```python
 # Standardisation
from sklearn.preprocessing import StandardScaler
 
 #crate instance of Scaler 
scale=StandardScaler()
 
 #Fit data to scale
scale.fit(x_train)

 #Tranform using scale
x_train_scaled=scale.transform(x_train)
x_test_scaled=scale.transform(x_test)
```

```python
 #Build Model
from sklearn.linear_model import LinearRegression

 # Create instance of model 
model=LinearRegression()
 #Fit model to training data
model.fit(x_train_scaled,y_train)
```

```python
 # coefficient and intercept

gradient = model.coef_[0]
intercept = model.intercept_

 # Print the coefficients.
print("gradient: ", gradient)
print("intercept: ", intercept)
```

```python
 # Evaluate model

pred=model.predict(x_test_scaled)

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

 #Mean absolute error
mae=mean_absolute_error(y_test,pred)
 #Mean squared error
mse=mean_squared_error(y_test,pred)
 # Root mean squared error
rmse=np.sqrt(mse)
 # Coefficient of determination
r2=r2_score(y_test,pred)

print("Mean absolute error: ",mae)
print("Mean squared error: ",mse)
print("Root mean squared error: ",rmse)
print("Coefficient of determination: ",r2)
```

```python
 # Line of best fit
x_fit=np.arange(0,25,1).reshape(-1,1)
x_fit_scaled=scale.transform(x_fit)
y_fit=model.predict(x_fit_scaled)


sns.scatterplot(data=video_df_new,x="NA_Sales",y="Global_Sales")
plt.plot(x_fit,y_fit,"red")
plt.xlim(-2,25)
plt.ylim(-2,35)
plt.xlabel("Sales of Video Game in North America")
plt.ylabel("Global Sales")
plt.title("Line of best fit of Global Sales vs Sales in North America")

#plt.savefig("LOFB_Global_sales_vs_NA_Sales.png")
plt.show()
```

### POLYNOMIAL REGRESSION

#### 1. Global Sales vs Year of release

```python
 # Extract Features
x4=video_df_new["Year_of_Release"]
y4=video_df_new["Global_Sales"]

 # Reshape x to have 2 dimensions
x4=x4.to_numpy().reshape(-1,1)

 # Split data into training and testing sets
x4_train,x4_test,y4_train,y4_test=train_test_split(x4,y4,test_size=0.2,random_state=42)
```

```python
 # Create an instance of the Standard Scaler
scale=StandardScaler()
 # Fit the standard scaler to input data
scale.fit(x4_train)

 # Apply standard scaler transformation to training and testing data
x4_train_scaled=scale.transform(x4_train)
x4_test_scaled=scale.transform(x4_test)

 # Import polynomial features 
from sklearn.preprocessing import PolynomialFeatures

 #Create an instance of Polynomial features
polynomial=PolynomialFeatures(degree=9,include_bias=False)
 #Apply polynomial features to training and testing datasets
x4_train_pol=polynomial.fit_transform(x4_train_scaled)
x4_test_pol=polynomial.fit_transform(x4_test_scaled)
```

```python
 # Create an instance of Linear regression model
model_poly=LinearRegression()
 # Fit model to the training data
model_poly.fit(x4_train_pol,y4_train)
```

```python
 # Get coefficients and intercept
c=model_poly.coef_
i=model_poly.intercept_

print("coefficients: ", c)
print("intercept: ", i)
```

```python
 # Evaluate model

 # get predicted values using test values from polynomial features
pred000=model_poly.predict(x4_test_pol)


mae4=mean_absolute_error(y4_test,pred000)
mse4=mean_squared_error(y4_test,pred000)
rmse4=np.sqrt(mse4)
r2_4=r2_score(y4_test,pred000)

print("Mean absolute error: ",mae4)
print("Mean squared error: ",mse4)
print("Root mean squared error: ",rmse4)
print("Coefficient of determination: ",r2_4)
```

```python
x_fit4=np.arange(1985,2017,1).reshape(-1,1)
x_fit4_scaled=scale.transform(x_fit4)
x_fit4_pol=polynomial.fit_transform(x_fit4_scaled)

y_fit4=model_poly.predict(x_fit4_pol)

sns.scatterplot(data=video_df_new,x="Year_of_Release",y="Global_Sales")
plt.plot(x_fit4,y_fit4,"red")
plt.ylim(-5,89)


plt.xlabel("Year of Release of Video Game")
plt.ylabel("Global Sales")
plt.title("Line of best fit of Global Sales vs Year of Release")

#plt.savefig("5.LOFB_Global_sales_vs_Year of release_deg=9.png")
plt.show()
```

```python
 # Import LabelEncoder class
from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()
 # Fit LabelEncoder to all categorical features
label.fit(x1_cat)
x1_cat_label=label.transform(x1_cat)

label.fit(x2_cat)
x2_cat_label=label.transform(x2_cat)

label.fit(x3_cat)
x3_cat_label=label.transform(x3_cat)

label.fit(x4_cat)
x4_cat_label=label.transform(x4_cat)

label.fit(x5_cat)
x5_cat_label=label.transform(x5_cat)

label.fit(x6_cat)
x6_cat_label=label.transform(x6_cat)

label.fit(x7_cat)
x7_cat_label=label.transform(x7_cat)
```

### MODEL 1: 3 Dense layers, dropout=20% layer, adam optimzer(Learning rate=0.001) default

```python
# Create an instance of the Sequential Class(Model)
from keras.models import Sequential
 # Import fully connected dense layers,Dropout
from keras.layers import Dense, Dropout, Input

a_mod=Sequential()
 #First hidden layer with 15 input dimensions, 64 units and ReLU activation function
a_mod.add(Input(shape=(15,)))  

a_mod.add(Dense(units = 64, activation = "relu"))
a_mod.add(Dropout(0.2))

 # Second hidden layer
a_mod.add(Dense(units = 64, activation = "relu"))


 #Third Hidden layer
a_mod.add(Dense(units = 64, activation = "relu"))


 # Output layer
a_mod.add(Dense(units = 1, activation = "linear"))

# Print a summary of the model
a_mod.summary()
plt.savefig("ANN1.png")
```

```python
 # Compile the model, using the default Adam optimizer
 #For loss function
from keras.losses import mean_squared_error
 # For evaluate model's performance
from keras.metrics import mean_squared_error

a_mod.compile(optimizer="adam", loss='mean_squared_error', metrics= ['mean_squared_error'])
```

```python
#Train the model
# Import the EarlyStopping class from Keras
from keras.callbacks import EarlyStopping

 # Early stopping criteria with patience of 20.
early_stopping = EarlyStopping(monitor='val_loss', patience = 20)

 # Train the model
a_history = a_mod.fit(x = all_x_train, y = all_y_train, batch_size = None, epochs = 200,
verbose = "auto", validation_split = 0.1, callbacks = [early_stopping] )
```

### k-Means

#### Global Sales and Sales in North America

```python
ft_names = ["NA_Sales","Global_Sales"]
x_k = video_df_new[ft_names]
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
scale.fit(x_k)

x_k_scaled = scale.transform(x_k)
```

```python
from sklearn.cluster import KMeans

inertias=[]
for k in range(1,30):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    kmeans.fit(x_k_scaled)

    inertia = kmeans.inertia_
    inertias.append(inertia)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6)) 

plt.plot(range(1, 30), inertias, marker='o')  #marker shows the plotted points
plt.title("Inertia for different values of k") 

plt.grid(True)  
plt.xticks(range(1,30))

plt.xlabel('Number of Clusters (k)') 
plt.ylabel('Inertia') 
plt.savefig("Innertia1.png")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 6, random_state = 42)
kmeans.fit(x_k_scaled)

inertia = kmeans.inertia_
inertia
```

```python
### Evaluation

cluster_labels_pred = kmeans.predict(x_k_scaled)
cluster_labels_pred
```

```python
from sklearn.metrics import davies_bouldin_score, silhouette_score

db_score = davies_bouldin_score(x_k_scaled, cluster_labels_pred)
print(f"DB score: {db_score}") # The smaller the beter

s_score = silhouette_score(x_k_scaled, cluster_labels_pred)
print(f"Silhouette score: {s_score}") 
# ranges from -1 to 1[closer to 1 means good clustering,0 means values close to boundaries,-1 means values in another cluster]
```

```python
%matplotlib inline
d=video_df_new[["Global_Sales","NA_Sales"]]
n_data=pd.DataFrame(d)

n_data["cluster"]=cluster_labels_pred  # includes cluster_labels_pred into dataframe as cluster
                                             # attached cluster_labels_predicted values to DataFrame

plot7=sns.scatterplot(data=n_data, x='NA_Sales', y='Global_Sales', hue='cluster',palette="viridis")
plt.xlabel("Sales in North America")
plt.ylabel("Global_Sales")
plt.xlim(-1,20)
plt.ylim(-1,40)
plt.title("Global Sales (y-axis) versus Sales in North America (x-axis)")
plt.legend(title='Cluster key',loc="lower right")
plt.savefig("k-means1.png")
plt.show()
```
