## Chi-sqare Test of Independence



### 1. Sex and Survive

**Null Hypothesis:** Sex and Survived are independent (no association).

**Alternative Hypothesis:** Sex and Survived are not independent (there is an association).

```python
cross_tab = pd.crosstab(df1['sex'], df1['survived'], margins=True)
print(cross_tab)
```

```python
from scipy.stats import chi2_contingency

cross_tab = pd.crosstab(df1['sex'], df1['survived'])

chi2, p, dof, expected = chi2_contingency(cross_tab)

print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)
```

We reject the null hypothesis of no association.

Chi-square statistic: A measure of how much the observed frequencies deviate from the expected frequencies.

p-value: The probability of observing the data if the null hypothesis is true. A lower p-value indicates a stronger association.

Degrees of freedom: The number of independent values that can vary in the calculation.

Expected frequencies: The frequencies expected under the null hypothesis of no association.

```python
 # Calculate survival rates by gender
survival_rates = df2.groupby('sex_male')['survived'].mean()
print(survival_rates)
```

This shows that males(True) had a 20% of survival as compared to females(False).

Females are more likely to survive than males.

### 2. Age and Survive

**Null Hypothesis:** Age and Survived are independent (no association).

**Alternative Hypothesis:** Age and Survived are not independent (there is an association).

```python
# Create age groups
df1['AgeGroup'] = pd.cut(df1['age'], bins=[0, 12, 18, 60, 100], labels=['Child', 'Teen', 'Adult', 'Senior'])

cross_tab_age = pd.crosstab(df1['AgeGroup'], df1['survived'], margins=True)
print(cross_tab_age)
```

```python
cross_tab_age = pd.crosstab(df1['age'], df1['survived'])

chi2_age, p_age, dof_age, expected_age = chi2_contingency(cross_tab_age)

print(f"Chi-square statistic: {chi2_age}")
print(f"p-value: {p_age}")
print(f"Degrees of freedom: {dof_age}")
print("Expected frequencies:")
#print(expected_age)
```

```python
We reject the null hypothesis of no association.
```

```python

# Create age groups
df2['AgeGroup'] = pd.cut(df2['age'], bins=[0, 12, 18, 60, 100], labels=['Child', 'Teen', 'Adult', 'Senior'])

# Calculate survival rates by age group
survival_rates = df2.groupby('AgeGroup')['survived'].mean()
print(survival_rates)
```

This shows that children and teenagers had a higher chance of survival than the adults.

`pd.cut`: This function from the pandas library is used to segment and sort data values into bins or categories.

`df1['age']`: This specifies the column in your DataFrame (df1) that contains the age values you want to categorize.

`bins=[0, 12, 18, 60, 100]`: This defines the edges of the bins. In this case, the bins are:

* 0 to 12 years (Child)

* 12 to 18 years (Teen)

* 18 to 60 years (Adult)

* 60 to 100 years (Senior)

`labels=['Child', 'Teen', 'Adult', 'Senior']`: These are the labels assigned to each bin. Each age value in the age column will be categorized into one of these labels based on the bin it falls into.

`df1['AgeGroup']`: This creates a new column in your DataFrame called AgeGroup, which contains the age group labels for each individual based on their age.
